# Normalize diamond sword textures

This notebook loads the sword textures from `data/sword`, verifies their dimensions, and writes 32x32 versions to `data/sword_32` when needed.

```python
from pathlib import Path
import subprocess
import sys

try:
    from PIL import Image
except ImportError:
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'pillow'])
    from PIL import Image

try:
    from IPython.display import display
except Exception:
    display = None
```

```python
SOURCE_DIR = Path(r"D:\alex's folder\code\mcproj\data\sword")
OUTPUT_DIR = Path(r"D:\alex's folder\code\mcproj\data\sword_32")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

image_files = sorted(p for p in SOURCE_DIR.glob('*.png') if p.is_file())
print(f'Found {len(image_files)} PNG files in {SOURCE_DIR}')
```

```python
results = []
target_size = (32, 32)

resample = getattr(Image, 'Resampling', Image).LANCZOS

for source_path in image_files:
    with Image.open(source_path) as image:
        original_size = image.size
        output_path = OUTPUT_DIR / source_path.name

        if original_size != target_size:
            normalized = image.convert('RGBA').resize(target_size, resample=resample)
            normalized.save(output_path)
            action = f'resized {original_size} -> {target_size}'
        else:
            image.save(output_path)
            action = 'already 32x32'

        results.append({
            'file': source_path.name,
            'original_size': original_size,
            'output_path': output_path,
            'action': action,
        })

for row in results:
    print(f"{row['file']}: {row['action']}")
```

```python
normalized_files = sorted(OUTPUT_DIR.glob('*.png'))
bad_files = []

for output_path in normalized_files:
    with Image.open(output_path) as image:
        if image.size != target_size:
            bad_files.append((output_path.name, image.size))

assert not bad_files, f'Non-32x32 images found: {bad_files}'
print(f'Verified {len(normalized_files)} files at 32x32 in {OUTPUT_DIR}')
```
